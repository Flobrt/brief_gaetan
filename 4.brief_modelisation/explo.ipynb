```python
# Colonnes commencant par ID 
list_id = []
for i in df.columns:
    if i.startswith('id'):
        print(i)
        list_id.append(i)
```

```python
mes_listes = {}
somme = 0

for i in list_id:
    print('='*50 + i + '='*50)
    columns = []
    for j in df.columns:
        for k in i.split('_')[1:]:
            if k in j:
                if j not in columns:
                    print(j)
                    columns.append(j)
                    somme += 1 

    mes_listes[i] = columns

print(f'Somme : {somme}')
```

```python
# Filtrer les colonnes qui ne commencent pas par "nb"
df = df.loc[:, ~df.columns.str.startswith('nb')]
```

```python
# df_client 
df_client = df[df['niveau'] == 'client']
# Supprimer des colonnes qui contiennent que des 0 ou des NaN
df_client = df_client.loc[:, ~(df_client == 0).all() & ~df_client.isna().all()]
df_client = df_client.loc[:, df_client.nunique(dropna=False) > 1]
df_client
```

```python
# df_contrat 
df_contrat = df[df['niveau'] == 'contrat']
# Supprimer des colonnes qui contiennent que des 0 ou des NaN
df_contrat = df_contrat.loc[:, ~(df_contrat == 0).all() & ~df_contrat.isna().all()]
df_contrat = df_contrat.loc[:, df_contrat.nunique(dropna=False) > 1]
df_contrat
```

```python
from sqlalchemy import create_engine, Integer, Date, Float, String

# Création des tables 
client = df['cd_client'].drop_duplicates()
client_columns = {
    "cd_client"  : String(20),
}

logement = df[['id_logement', 'cd_type_habitat_logement', 'lb_type_habitat_logement']].drop_duplicates()
logement_columns = {
    'id_logement'                   : String(20),       
    'cd_type_habitat_logement'      : String(20),
    'lb_type_habitat_logement'      : String(255),
}

souscription = df[['id_produit_contrat', 'cd_produit_contrat', 'grp_produit_contrat', 'lst_cd_produit_combine_contrat', 'fg_contrat_acquis',
                   'fg_contrat_primo_acquis', 'fg_contrat_der_acquis', 'fg_contrat_acquis_partenaire', 'fg_rib_renseigne_pre_contrat']].drop_duplicates()
souscription_columns = {
    'id_produit_contrat'                       : String(255),
    'cd_produit_contrat'                       : String(255), 
    'grp_produit_contrat'                      : String(255), 
    'lst_cd_produit_combine_contrat'           : String(255), 
    'fg_contrat_acquis'                        : String(255),
    'fg_contrat_primo_acquis'                  : String(255), 
    'fg_contrat_der_acquis'                    : String(255), 
    'fg_contrat_acquis_partenaire'             : String(255), 
    'fg_rib_renseigne_pre_contrat'             : String(255),
}

campagne_offre_mkt = df[['id_campagne_souscrit', 'id_campagne_offre_mkt']].drop_duplicates()
campagne_offre_mkt_columns = {
    "id_campagne_souscrit"     : String(255),
    "id_campagne_offre_mkt"    : String(255),
}

vendeur = df[['id_agent_vendeur', 'id_vendeur_partenaire']].drop_duplicates()
vendeurt_columns = {
    "id_agent_vendeur"         : String(255),
    "id_vendeur_partenaire"    : String(255),
}

# contrat = df['id_contrat', 'cd_contrat', 'version_contrat', 'version_contrat_r2', 'version_contrat_r3', 'version_contrat_r5'].drop_duplicates()
# contrat_columns = {
#     'id_contrat'            : String(255),
#     'cd_contrat'            : String(255),
#     'version_contrat'       : String(255),
#     'version_contrat_r2'    : String(255),
#     'version_contrat_r3'    : String(255),
#     'version_contrat_r5'    : String(255),
# }

assureur = df[['cd_assureur_contrat', 'lb_assureur_contrat']].drop_duplicates()
assureur_columns = {
    'cd_assureur_contrat'   : String(255),
    'lb_assureur_contrat'   : String(255),
}

marque = df[['id_marque', 'cd_marque']].drop_duplicates()
marque_columns = {
    'id_marque'             : String(255),
    'cd_marque'             : String(255),
}


engine = create_engine('mysql+pymysql://root:password@127.0.0.1:3306/speed')

client.to_sql('client', con=engine, if_exists='replace', index=False, dtype=client_columns)
logement.to_sql('logement', con=engine, if_exists='replace', index=False, dtype=logement_columns)
souscription.to_sql('souscription', con=engine, if_exists='replace', index=False, dtype=souscription_columns)
campagne_offre_mkt.to_sql('campagne_offre_mkt', con=engine, if_exists='replace', index=False, dtype=campagne_offre_mkt_columns)
vendeur.to_sql('vendeur', con=engine, if_exists='replace', index=False, dtype=vendeurt_columns)
# contrat.to_sql('contrat', con=engine, if_exists='replace', index=False, dtype=contrat_columns)
assureur.to_sql('assureur', con=engine, if_exists='replace', index=False, dtype=assureur_columns)
```
