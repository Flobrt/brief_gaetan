```python
import pandas as pd
from sqlalchemy import create_engine, Integer, String, Float, Date
import ftfy
from unidecode import unidecode
```

```python
pd.set_option('display.max_columns', 500)
```

```python
df_1 = pd.read_excel('data/donnees-communes.xlsx', engine='openpyxl')
df_1
```

```python
df_1 = df_1[df_1['COM'].apply(lambda x: not isinstance(x, bool))]
bool_values_df = df_1[df_1['COM'].apply(lambda x: isinstance(x, bool))]
bool_values_df
```

```python
# reformatage des communes 
df_1['COM'] = df_1['COM'].apply(ftfy.fix_text)
df_1['COM'] = df_1['COM'].apply(unidecode)
df_1['COM'] = df_1['COM'].str.upper()
df_1['code_clean'] = df_1['CODDEP'].astype(str) + df_1['COM'].astype(str)
df_1
```

```python
df_1 = df_1.rename(columns={'CODDEP' : 'dep_code'})
```

```python
df_2 = pd.read_excel('data/fr-esr-referentiel-geographique.xlsx')
df_2 = df_2[['reg_code', 'reg_nom', 'dep_code']].drop_duplicates()
df_2
```

```python
df_3 = pd.merge(df_2, df_1, on='dep_code', how='inner')
df_3["code_merge"] = df_3['dep_code'].astype(str) + df_3['CODCOM'].astype(str)
df_3
```

```python
result = df_3[df_3['code_merge'] == '1143']
result
```

```python
df_4 = pd.read_excel('data/valeurs-foncieres.xlsx')
df_4['id_vente'] = range(1, len(df_4) + 1)
df_4['Code departement'] = df_4['Code departement'].astype(str)
df_4['concatcode'] = df_4['No voie'].astype(str) + df_4['Voie'].astype(str) + df_4['Code postal'].astype(str)
df_4['id_bien'] = pd.factorize(df_4['concatcode'])[0]+1
df_4['code_merge'] = df_4['Code departement'].astype(str) + df_4['Code commune'].astype(str)
df_4['code_clean'] = df_4['Code departement'].astype(str) + df_4['Commune'].astype(str)
df_4
```

```python
# df_5
df_5 = pd.merge(df_3, df_4, on='code_merge', how='inner')
df_5
```

```python
df_6 = pd.merge(df_3, df_4, on='code_clean', how='inner')
df_6
```

```python
# Itérer sur les colonnes de type 'object' (chaînes de caractères)
for col in df_5.select_dtypes(include='object').columns:
    # Appliquer ftfy, unidecode et str.upper seulement si la valeur est une chaîne
    df_5[col] = df_5[col].apply(lambda x: ftfy.fix_text(str(x)) if isinstance(x, str) else str(x))
    df_5[col] = df_5[col].apply(lambda x: unidecode(x) if isinstance(x, str) else x)
    df_5[col] = df_5[col].apply(lambda x: x.upper() if isinstance(x, str) else x)

df_5
```

```python
ventes = df_4[['id_vente', 'Date mutation', 'Valeur fonciere', 'id_bien']].drop_duplicates()
ventes_columns = {
    "id_vente"  : Integer(),
    "Date"      : Date(),
    "Valeur"    : Integer(),
    "id_bien"   : Integer()
}

biens = df_4[['id_bien', 'No voie', 'Type de voie', 'Voie', 'Nombre pieces principales', 'Surface Carrez du 1er lot', 'Surface reelle bati', 'Type local', 'Code ID commune']].drop_duplicates()
biens_columns = {
    "id_bien"       : Integer(),
    "No_voie"       : Integer(),
    "Type_voie"     : String(4),
    "Voie"          : String(50),
    "Total_piece"   : Integer(),
    "Surface_carrez": Float(),
    "Surface_local" : Float(),
    "Type_local"    : String(50),
    "Id_commune"    : Integer(),
}


commmunes = df_3[['CODCOM', 'COM', 'dep_code', 'CODREG', 'PTOT']].drop_duplicates()
commmunes_columns = {
    "id_commune"        : Integer(),
    "Nom_commune"       : String(255),
    "Code_departement"  : String(255),
    "id_region"         : String(255),
    "Population"        : String(255)
}


region = df_5[['CODREG', 'reg_nom']].drop_duplicates()
region_columns = {
    "id_region"     : Integer(),
    "nom_region"    : String(50),
}



engine = create_engine('mysql+pymysql://root:password@127.0.0.1:3306/immo')


ventes.to_sql('Ventes', engine, if_exists='replace', index=False, dtype=ventes_columns)
biens.to_sql('Biens', engine, if_exists='replace', index=False, dtype=biens_columns)
commmunes.to_sql('Communes', engine, if_exists='replace', index=False, dtype=commmunes_columns)
region.to_sql('Regions', engine, if_exists='replace', index=False, dtype=region_columns)
```
