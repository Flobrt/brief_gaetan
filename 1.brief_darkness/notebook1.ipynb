```python
import pandas as pd
```

```python
df = pd.read_csv('data/darkness-dataset-671215d22f07d453948548.csv', sep=';///', encoding='utf-8', engine='python')
df
```

```python
df['id,name,country_code,position,city,current_company,about,posts,experience,url,people_also_viewed,educations_details,education,recommendations_count,avatar,languages,certifications,recommendations,volunteer_experience,courses,followers,connections,current_company_company_id,current_company_name,publications,patents,projects,organizations,input_url,linkedin_id;;;;'] = df['id,name,country_code,position,city,current_company,about,posts,experience,url,people_also_viewed,educations_details,education,recommendations_count,avatar,languages,certifications,recommendations,volunteer_experience,courses,followers,connections,current_company_company_id,current_company_name,publications,patents,projects,organizations,input_url,linkedin_id;;;;'].str.replace('&amp;', '&')
```

```python

list_ligne = []

for i in df['id,name,country_code,position,city,current_company,about,posts,experience,url,people_also_viewed,educations_details,education,recommendations_count,avatar,languages,certifications,recommendations,volunteer_experience,courses,followers,connections,current_company_company_id,current_company_name,publications,patents,projects,organizations,input_url,linkedin_id;;;;'].unique():
    nouvelle_chaine = ""
    a = 0 
    start = 0 
    last = ""
    last_last = ""

    for j in i: 
        if start == 0 and j == ',':
            nouvelle_chaine += ';;'
            start = 1
            continue
        if j == ',' and a == 0 and last == '"' and last_last == '"':
            nouvelle_chaine += ';;'
        elif j == '{' or j == '[':
            a += 1
            nouvelle_chaine += j
        elif j == '}' or j == ']':
            a -= 1
            nouvelle_chaine += j
        else:
            nouvelle_chaine += j
        last_last = last
        last = j

        # if a == 0:
        #     if len(nouvelle_chaine) > 2:
        #         if j == '"' and last == '"' and last_last == ',':
        #             nouvelle_chaine = nouvelle_chaine[:-2]
        #             nouvelle_chaine += ';;'
        #             nouvelle_chaine += '"'
        #             nouvelle_chaine += '"'

    
    list_ligne.append(nouvelle_chaine)
```

```python
columns = df.columns[0].replace(';',',').split(',')
print(columns)
print(len(columns))
```

```python
data_split = [ligne.split(';;') for ligne in list_ligne]
nb = 1
for i in data_split[:-1]:
    print(len(i))
    
```

```python
df = pd.DataFrame(data_split, columns=columns[:-3])
```
