```python
# Pays ayant gagné le plus de matchs en coupe du monde df['winner'] prends la valeur de la colonne home_team si home_result > away_result sinon away_team
df_winner = df.copy()

def qui_gagne(x):
    if x['home_result'] > x['away_result']:
        return x['home_team']
    elif x['home_result'] == x['away_result']:
        return 'draw'
    else:
        return x['away_team']

df_winner['winner'] = df.apply(qui_gagne, axis=1)
df_winner['winner'].value_counts()
```

```python
# Le plus grand ecart de buts dans un match 
df['diff'] = abs(df['home_result'].astype(int) - df['away_result'].astype(int))
df['diff'].sort_values(ascending=False)
```

```python
# combien de matchs nuls au total ?
df_winner['winner'] = df_winner.apply(lambda x: 'draw' if x['home_result'] == x['away_result'] else x['winner'], axis=1)
df_winner['winner'].value_counts()['draw']
```
